```python
fields_success = df.groupby('Field')['Success_Rate_From'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
fields_success.plot(kind="bar" , color="blue")
plt.title("Fields Success")
plt.xlabel("Fields")
plt.ylabel("Number of Successes")
plt.xticks(rotation=45)
plt.show()
```
